```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import openpyxl
```

Gr√°fico de ingresos mensuales

```python
# Lectura del archivo
monthly_sales = pd.read_excel('./data/ingresos_mensuales.xlsx')

# Crear gr√°fico
plt.subplots(figsize=(12, 6))
plt.bar(monthly_sales['mes'], monthly_sales['ingresos_mensuales'])

# Ciclo para agregar etiquetas con valores a las barras
for i, (mes, ingreso) in enumerate(zip(monthly_sales['mes'], monthly_sales['ingresos_mensuales'])):
    plt.text(mes, ingreso, f'{ingreso:,.2f}', ha='center', va='bottom', fontsize=9, color='red')

# Personalizaci√≥n del gr√°fico
plt.title('Ingresos Mensuales en 2023', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ingresos Mensuales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Gr√°fico de ventas por departamento

```python
# Cargar los datos
sales_departmet = pd.read_excel('./data/ventas_departamento.xlsx')

# Convertir los ID de departamento a texto
sales_departmet['id_departamento'] = sales_departmet['id_departamento'].astype(str)

# Crear gr√°fico de dona
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(
    sales_departmet['total_ventas'],
    labels=sales_departmet['id_departamento'],
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    wedgeprops={'width': 0.4},
    textprops={'fontsize': 10}
)

# C√≠rculo blanco en el centro (efecto de dona)
plt.gca().add_artist(plt.Circle((0, 0), 0.6, color='white'))

# T√≠tulo
plt.title('Total Ventas por Departamento 2023', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

Gr√°fico de distribuci√≥n de las ventas

```python
distribution = pd.read_excel('./data/distribucion_ventas.xlsx')

# Convertir los ID de departamento a texto
distribution['id_seccion'] = distribution['id_seccion'].astype(str)

# Crear √≠ndices consecutivos para alinear las barras una al lado de la otra
indices = range(len(distribution))

# Crear gr√°fico
plt.figure(figsize=(10, 6))
plt.bar(indices, distribution['distribucion'])

# Agregar etiquetas de texto sobre las barras
for i, ventas in enumerate(distribution['distribucion']):
    plt.text(i, ventas, f'{ventas:,.2f}', ha='center', va='bottom', fontsize=9, color='red')

# Configurar etiquetas del eje X como texto
plt.xticks(indices, distribution['id_seccion'])

# T√≠tulos y etiquetas
plt.title('Distribuci√≥n de Ventas 2023', fontsize=16)
plt.xlabel('Secci√≥n', fontsize=12)
plt.ylabel('Ventas Totales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Gr√°fica Cantidades vendidas por producto

```python
# Leer archivo
sales_product = pd.read_excel('./data/ventas_producto.xlsx')

# Crear gr√°fico
plt.figure(figsize=(12, 8))
plt.barh(sales_product['nombre_producto'], sales_product['total_vendido'], color='skyblue')

# Agregar etiquetas con los valores en las barras
for i, (producto, valor) in enumerate(zip(sales_product['nombre_producto'], sales_product['total_vendido'])):
    plt.text(valor + 1000, i, f'{valor:,}', va='center', fontsize=9, color='black')  # ajusta +1000 seg√∫n escala

# Personalizaci√≥n
plt.title('Cantidades Vendidas por Producto 2023', fontsize=16)
plt.xlabel('Cantidad Vendida', fontsize=12)
plt.ylabel('Producto', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6, axis='x')

plt.tight_layout()
plt.show()
```

Gr√°fica ingresos por producto

```python
# Leer archivo
revenue_product = pd.read_excel('./data/ingresos_producto.xlsx')

# Ordenar datos de mayor a menor resultado
revenue_product = revenue_product.sort_values('total_vendido')

# Crear gr√°fico
plt.figure(figsize=(12, 8))
plt.plot(revenue_product['total_vendido'], revenue_product['nombre_producto'], marker='o', linestyle='-', color='teal')

# Agregar etiquetas de valor al lado de cada punto
for x, y in zip(revenue_product['total_vendido'], revenue_product['nombre_producto']):
    plt.text(x + 1000, y, f'{x:,.2f}', va='center', fontsize=9, color='red')

# Personalizaci√≥n de gr√°fico
plt.title('Ingresos por Producto 2023', fontsize=16)
plt.xlabel('Total Vendido', fontsize=12)
plt.ylabel('Producto', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6, axis='x')

plt.tight_layout()
plt.show()
```

Gr√°fica de Ventas por Cliente

```python
# Leer datos
customer_sales = pd.read_excel('./data/ventas_clientes.xlsx')

# Ordenar de mayor a menor por total_compras
ordered_sales = customer_sales.sort_values('total_compras', ascending=True)

# Crear gr√°fico de barras horizontales
plt.figure(figsize=(10, 8))
plt.barh(ordered_sales['id_cliente'].astype(str), ordered_sales['total_compras'], color='skyblue')

# Agregar etiquetas
for i, (compra, cliente) in enumerate(zip(ordered_sales['total_compras'], ordered_sales['id_cliente'])):
    plt.text(compra + 20, i, f'{compra:,.2f}', va='center', fontsize=8)

# Personalizaci√≥n
plt.title('Top 20 Clientes por Total de Compras', fontsize=16)
plt.xlabel('Total Compras')
plt.ylabel('ID Cliente')
plt.grid(True, axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
